#Parte #1

```
import matplotlib.pyplot as plt
import numpy as np

Tliq = 25
Tman = 130
r1 = 1
r2 = 1.7
r = np.linspace(1,1.7,20)

y = Tliq + ((Tliq-Tman)/((1/r1) - (1/r2)))*((1/r) - (1/r1))
```

## Perfil de Temperatura: 
**T(r) =**
$$
   Tliq + (Tliq-Tman)\over(1/r1 - 1/r2)(1/r - 1/r1)
$$

```
plt.plot(r,y)
plt.title('Perfil de Temperatura')
plt.xlabel('Raio(m)')
plt.ylabel('Temperatura(CÂ°)')
plt.savefig('Grafico.png',dpi= 300)
```

##VariÃ¡veis

```
k=0.79  # W/mK
h=630   # W/(m2 K)
Tinf = 25
Tbase =130
r = 1
w = 0.25
t = 0.142
L = 0.33
```

##L(estendido), PerÃ­metro e Ãrea

```
Lc = L + t/2 
P = 2*w + 2*t
A = w*t
```

```
m = Tbase * np.sqrt(h*P*k*A)
m
```

##Definimos um espaÃ§o em **X** variando de 0 atÃ© o **L** escolhido

Note que no nosso exemplo, ao colocarmos o **X** como sendo o nosso valor final de 0.33m o grÃ¡fico apresenta um comportamento atÃ­pico, mostranque que a troca de calor na aleta Ã© tal qual como se nÃ£o houvesse a mesma. 

```
# np.linspace(inÃ­cio, fim, de quanto em quanto)
x = np.linspace(0,0.05,50)
x
```

###Agora calculamos o perfil de temperatura

```
y = 130*(np.cosh(m*(Lc-x))/np.cosh(m*Lc))
y
```

Considerando o raio variando atÃ© 0,05 metros, podemos ver o comportamento da aleta

```
plt.plot(x,y)
plt.xlabel('Comprimento da Aleta(m)')
plt.ylabel('Temperatura(CÂ°)')
plt.savefig('Grafico.png',dpi= 300)
```

Agora demonstrando para L =  0.33m

```
# np.linspace(inÃ­cio, fim, de quanto em quanto)
x_2 = np.linspace(0,0.33,100)
y_2 = 130*(np.cosh(m*(Lc-x_2))/np.cosh(m*Lc))

plt.plot(x_2,y_2)
plt.xlabel('Comprimento da Aleta(m)')
plt.ylabel('Temperatura(CÂ°)')
plt.savefig('Grafico2.png',dpi= 300)
```
